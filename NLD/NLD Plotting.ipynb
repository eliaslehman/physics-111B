```python
import os
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
folder_path = "/Users/elias/Desktop/School/physics-111B/NLD/data/pn-bifur"

def plot_files_in_folder(folder_path):
    # Get a list of all files in the folder
    files = os.listdir(folder_path)

    # Iterate through each file in the folder
    for file_name in files:
        file_path = os.path.join(folder_path, file_name)

        # Check if the item in the folder is a file (not a subfolder)
        if os.path.isfile(file_path) and file_name.endswith('.csv'):
            # Read data from the CSV file
            with open(file_path, 'r') as csv_file:
                csv_reader = csv.reader(csv_file)
                # Extract time and signal data
                data = list(csv_reader)
                time = [float(row[0]) for row in data]
                signal = [float(row[1]) for row in data]

            # Plot the data
            plt.scatter(time, signal, label=file_name, marker ='.', s = 1)
            # Add labels and legend
            plt.xlabel('Driving Voltage Magnitude (V)')
            plt.ylabel('Output Current (V)')
            plt.legend()
            plt.title('PN Junction Current vs Driving Voltage')
            plt.grid()
            # Show the plot
            plt.show()

# Call the function with the specified folder path
plot_files_in_folder(folder_path)
```
